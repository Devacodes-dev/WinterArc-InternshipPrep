```python
folder_name='star_plots'
os.makedirs(folder_name,exist_ok=True)  # use to create a directory to store our visualization images
base_dir=f'/content/{folder_name}/'
```

1) VISUALIZE THE COUNT OF STARS TYPES

```python
df['Star type'].value_counts().plot(kind="bar")
plt.show()
```

2) customize(les make it pretty understanding) and save it

```python
plt.figure(figsize=(5,5))
plt.title('Visualize star count per star type', color='purple')
ax=df['Star type'].value_counts().plot(kind="bar", color=['yellow','red','lightblue','pink','green','royalblue'])  # saved the plot in a variable
plt.style.use('Solarize_Light2')
ax.bar_label(ax.containers[0],color='purple') # display the count of each bar in the plot , color
plt.ylabel('number of stars', color='purple') # title for y axis and its color
plt.xlabel('star type',color='purple')        # title for x axis and its color
plt.yticks(color='magenta')                   #  give color to the intervals of y axis
plt.xticks(ticks=[0,1,2,3,4,5], labels=['Brown\nDwarf','Red\nDwarf','White\nDwarf' ,'Main\nSequence','Super\nGiant','Hyper\nGiant'], rotation=45,color='magenta')
plt.savefig(base_dir+'star_type.png')
plt.show()
```

```python
print(plt.style.available)     # command to get all the style names available
```

3) use seaborn to make the barplot and compare the outputs

```python
# df['Star color'].value_counts()  - this prints the output with two columns: color and its respective count
#df['Star color'].value_counts().index  # this prints only the colors which the index
ax =sns.barplot(x=df['Star color'].value_counts().index , y=df['Star color'].value_counts(),
            palette='pastel6')
for container in ax.containers:
    ax.bar_label(container,color='brown')
plt.title("Visualize the count of each star color", color='brown', fontsize=15,weight='heavy')
plt.xticks(rotation=15,color='brown')
plt.xlabel('star color',color='brown')
plt.ylabel('star count per color',color='brown',fontsize=13)
plt.yticks(color='brown')
plt.savefig(base_dir+'star_color.png')
plt.show()
```

VISUALIZE OUTLIERS


```python
plt.figure(figsize=(20,8))
plt.style.use('default')
plt.suptitle('Visualizing the outliers in numeric features using star type', color='black', fontsize=15, weight='bold')

for i in range(4):
  plt.subplot(1,4,i+1)
  sns.boxplot(x=df['Star type'],y=df.iloc[:,i],palette='pastel6')
  plt.ylabel('')
  plt.title(df.columns[i], color='red')

plt.savefig(base_dir+'boxplot_star_type.png')
plt.show()
```
