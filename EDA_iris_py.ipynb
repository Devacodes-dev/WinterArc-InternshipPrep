DATA VISUALIZATION USING MATHPLOTLIB AND SEABORN LIBRARY


```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x="Species",data=df)   # used to display count of each unique value as a x-y plot
plt.show()
```

```python
import seaborn as sns
sns.scatterplot(x="SepalLengthCm",y="SepalWidthCm",data=df)    # compares sepal length and width of all 3 tyes together
```

```python
sns.scatterplot(x="SepalLengthCm", y="SepalWidthCm", hue="Species", data=df)
```

```python
sns.pairplot(df.drop(['Id'], axis = 1) ,hue='Species',height=2)  #all columns's relationships using pairplot
```

```python
fig,axes=plt.subplots(2,2,figsize=(10,10))   # creating a 2 row and 2 col figure

axes[0,0].set_title("Sepal Length")  # it is used to set the title of the graph
axes[0,0].hist(df['SepalLengthCm'],bins=5)  # bins are used to range and divide the values of the specified column

axes[0,1].set_title("Sepal Width")
axes[0,1].hist(df['SepalWidthCm'],bins=5) # axes[0,1] refers to 0th row and 1st col

axes[1,0].set_title("Petal Length")
axes[1,0].hist(df['PetalLengthCm'],bins=5)

axes[1,1].set_title("Petal Width")
axes[1,1].hist(df['PetalWidthCm'],bins=5)
```

```python
''' histogram with distplot
plot=sns.FacetGrid(df,hue='Species')
plot.map(sns.distplot,'SepalWidthCm' ).add_legend()

these code has been update with histplot that gives same outputbut modern function '''

sns.histplot(df['SepalLengthCm'],bins=10,kde=True)
sns.histplot(df['PetalLengthCm'],bins=10,kde=True)
sns.histplot(df['SepalWidthCm'],bins=10,kde=True)
sns.histplot(df['PetalWidthCm'],bins=10,kde=True)   #overlapping issue to solve that use plt
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data=df,x='SepalLengthCm',kde=True,hue='Species')

plt.figure(figsize=(8,4))
sns.histplot(data=df,x='SepalWidthCm',kde=True, hue='Species')

plt.figure(figsize=(8,4))
sns.histplot(data=df,x='PetalLengthCm',kde=True,hue='Species')

plt.figure(figsize=(8,4))
sns.histplot(data=df,x='PetalWidthCm',kde=True, hue='Species')    # can check for overlapping and least can be used as classification feature
```
